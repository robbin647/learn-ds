## Introduction to Numpy and pandas

```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame

print(Series)
print(DataFrame)
```

```python
series_obj = Series(np.arange(8), index=['row1', 'row2', 'row3', 'row4', 
                                             'row5', 'row6', 'row7', 'row8'])
series_obj
```

```python
series_obj['row7'] # select an element from a series
```

```python
series_obj[[0,7]] # select a range of elements
```

```python
np.random.seed(11)
df_obj = DataFrame(np.random.rand(36).reshape((6,6)), 
                   index= ["row1", 'row2', 'row3', 'row4', 'row5', 'row6'],
                   columns = ['col1', 'col2', 'col3', 'col4', 'col5', 'col6'])
df_obj
```

```python
df_obj.loc[['row2', 'row5'], ['col2', 'col5']]
```

```python
series_obj['row2' : 'row7']
# df_obj['row2':'row5']['col2':'col5']
```

Comparison with scalar

```python
series_obj < 8
df_obj < 0.2
```

```python
series_obj[series_obj > 5]
df_obj.loc[['row1', 'row5'], ['col2', 'col5']] = 100
df_obj
```
